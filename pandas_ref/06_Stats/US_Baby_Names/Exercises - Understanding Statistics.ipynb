### Step 6. Is there more male or female names in the dataset?

```python
print(baby_names.Gender.value_counts())
baby_names.Gender.value_counts().plot(kind = 'bar')
```

> There are more Female babies than Male babies in our dataset

### Step 7. Group the dataset by name and assign to names

```python
names = baby_names.groupby(['Name']).sum()
```

### Step 8. How many different names exist in the dataset?

```python
names.head(1)
# We don't need the Year and its sum .. Let's drop that column tooo
names = names.drop(['Year'],axis=1)
```

```python
print(names.shape)
names.head(1)
```

```python
names.sort_values('Count',ascending = False)
```

> There are 17632 different Names in our dataset.. and Highly repeated name 'jacob' : 242874 times

### Step 11. What is the median name occurrence?

```python
print(len(names[names.Count == names.Count.median()]))
names[names.Count == names.Count.median()]
```
