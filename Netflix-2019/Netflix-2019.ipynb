```python
sns.set_style('darkgrid')
plt.figure(figsize=(8,12))
sns.countplot(df2.type)
plt.title('Movies and Tv-series available in Netflix')
```

* The Movies are more than TV-shows on Netflix in 2019
* lets check with release dates

```python
sns.set_style('darkgrid')
plt.figure(figsize=(16,12))
plot1 = sns.countplot(df2.release_year,hue=df2.type)
plt.title('Movies and Tv-series available in Netflix')
plt.legend(loc='upper left')
plot1.set_xticklabels(plot1.get_xticklabels(),rotation=45)
plt.show()
```

> Here we can observe that there is drastic increase in movies from 2014 into Netflix
