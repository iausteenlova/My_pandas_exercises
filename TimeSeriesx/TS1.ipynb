###### *  pd.date_range()

```python
# pd.date_range(start=None,end=None,periods=None,freq=None,tz=None,normalize=False,name=None,closed=None,**kwargs,)
```

```python
# Yearly end : 'A' ; Yearly start : "AS"
pd.date_range('2019-01-1',periods=10,freq='AS')

# Quarter end : 'Q' ; Business Quarter : "BQ" ; Quarter start : "QS"
pd.date_range('2019-01-1',periods=10,freq='Q')

# Month End : 'M' ; Month start : "MS" ; Semi-Month : "SM" & "SMS"
pd.date_range('2019-01-1',periods=10,freq='M')

# Daily : 'D'
pd.date_range('2019-01-1',periods=10,freq='D')

# Hour : "H"
pd.date_range('2019-01-1',periods=10,freq='H')

# Minutely : "T" or "min"
pd.date_range('2019-01-1',periods=10,freq='min',)

# Seconds : "S"
pd.date_range('2019-01-1',periods=10,freq='S',)

# milli seconds : "L" or "ms"
pd.date_range('2019-01-1',periods=10,freq='ms',)

# micro seconds : "U" or "us"
pd.date_range('2019-01-1',periods=10,freq='U',)

# micro seconds : "N"
pd.date_range('2019-01-1',periods=10,freq='N',)
```

###### >> Indexing with Time Series

```python
rng = pd.date_range('2016 Jul 1', periods = 10, freq = '15D')
rng
ys = pd.Series(range(len(rng)), index = rng)
print(type(ys.index))
ys
```

```python
# You can also use time period indices, in cases where it makes more sense 
# to think about your index as a time span rather than a single point in time

periods = [pd.Period('2016-01'), pd.Period('2016-02'), pd.Period('2016-03')]
ts = pd.Series(np.random.randn(len(periods)), index = periods)
print(type(ts.index))
ts
```

```python
ts['2016-03']
```

```python
ys['2016-07']
```

```python
us = pd.Series(range(10), pd.date_range('07-10-16 8:00', periods = 10, freq = 'H'))
print(type(us.index))
us
```

* Timestamped data can be converted to period indices with to_period and vice versa with to_timestamp

```python
us_period = us.to_period()
print(type(us_period.index))
us_period
```

```python
us_period['2016-07-10 08:30':'2016-07-10 11:45'] # we have the concept of overlap with time periods
```

```python
us['2016-07-10 08:30':'2016-07-10 11:45'] # we have the concept of include with timestamps
```
